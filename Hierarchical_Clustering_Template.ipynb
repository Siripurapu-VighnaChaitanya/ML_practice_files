# Hierarchical Clustering

## Importing libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing dataset

```
ds=pd.read_csv('Amazon.csv')
x=ds.iloc[:,[2,4]].values
print(x)
```

## Optimal number of clusters via DendoGram

```
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('age')
plt.ylabel('rating')
plt.show()
```

## Hierarchical Clustering Model Training on Training set

```
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=6,metric='euclidean',linkage='ward')
ahc=hc.fit_predict(x)
print(ahc)
```

## Visualizing Clusters

```
plt.scatter(x[ahc==0,0],x[ahc==0,1],label='cluster 0')
plt.scatter(x[ahc==1,0],x[ahc==1,1],label='cluster 1')
plt.scatter(x[ahc==2,0],x[ahc==2,1],label='cluster 2')
plt.scatter(x[ahc==3,0],x[ahc==3,1],label='cluster 3')
plt.scatter(x[ahc==4,0],x[ahc==4,1],label='cluster 4')
plt.scatter(x[ahc==5,0],x[ahc==5,1],label='cluster 5')
plt.title('Clusters of customers')
plt.xlabel('age')
plt.ylabel('rating')
plt.legend()
plt.show()
```
